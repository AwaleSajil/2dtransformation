```python
import numpy as np
from matplotlib import pyplot as plt
from math import sin, cos, radians
%matplotlib inline

#points = np.random.randint(low = -15, high= 15, size= (3, 2))
points = [[-1, 3, 0],
         [15, 5, 0],
         [10, 15, 0]]
points = np.vstack((points, points[0]))
pivot = np.array([0,0, 0])

shiftpoints = points - pivot
```

```python
#about x axis
Rx  = np.array([[1, 0],
               [0 ,-1],
               [0, 0]])

newpoints = np.dot(shiftpoints, Rx)

plt.plot(points[:,0], points[:,1])
plt.plot(newpoints[:,0], newpoints[:,1])
```

```python
#about y axis
Rx  = np.array([[-1, 0],
               [0 ,1],
               [0, 0]])

newpoints = np.dot(shiftpoints, Rx)

plt.plot(points[:,0], points[:,1])
plt.plot(newpoints[:,0], newpoints[:,1])
```

```python
#about origin

Rx  = np.array([[-1, 0],
               [0 ,-1],
               [0,0]])

newpoints = np.dot(shiftpoints, Rx)

plt.plot(points[:,0], points[:,1])
plt.plot(newpoints[:,0], newpoints[:,1])
```

```python
# about a line y = mx + c
m = 1
c = 10
    
Ref = np.array([[1-m*m, 2*m, -2*c*m],
                [2*m, m*m -1, 2*c],
               [0 , 0, 1*m*m]])

newpoints = np.dot(shiftpoints, Ref)

plt.plot(points[:,0], points[:,1])
plt.plot(newpoints[:,0], newpoints[:,1])
```
