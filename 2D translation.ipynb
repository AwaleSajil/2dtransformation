```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

points = np.random.randint(low = -15, high= 15, size= (3, 2))
points = np.vstack((points, points[0]))

translate = np.random.randint(low = -5, high = 5, size = (1,2))

newpoints = points + translate
```

```python
print("points = ", points)
print("translate = ", translate)
print("newpoints = ", newpoints)
```

```python
plt.plot(points[:,0], points[:,1])
plt.plot(newpoints[:,0], newpoints[:,1])
```
